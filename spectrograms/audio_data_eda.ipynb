# Audio Data EDA

In this notebook, I will be exploring the audio files to see what I can learn from them

```python
#getting all the polly generated audio files and the splices of podcast episodes
#the goal is explore the similarities and difference between our AWS Polly generated data and the splices of a podcast
#episode
import glob
import warnings
warnings.filterwarnings('ignore')

podcast_audio_files = glob.glob('./audio-split-data/*.mp3', recursive=True)
polly_audio_files = glob.glob('./polly-data/*.mp3', recursive=True)
```

```python
#sorting polly audio files
pa = polly_audio_files.copy()
ordered_pa = sorted(pa)
```

```python
#sorting podcast episode splices
pod_a = podcast_audio_files.copy()
ordered_pod = sorted(pod_a)
```

```python
#grouping the audio file together by word
grouped_audio = list(zip(ordered_pa, ordered_pod))
```

```python
#for loading and visualizing audio data
import librosa
from librosa import display
```

```python
#loading both the AWS polly generated audio and the actual episode audio chunks into librosa
import numpy as np

polly_episode_arr = []
for group in grouped_audio:
    polly = group[0]
    episode_chunks = group[1]
    episode_id = polly.split('/')[-1].split('.')[0]
    try:
        y_pol, sr_pol = librosa.load(polly, sr=11025)
        y_episode, sr_episode = librosa.load(episode_chunks)
        polly_episode_arr.append([episode_id, y_pol, y_episode])
    except Exception as e:
        print(group)
print(polly_episode_arr)
```

```python
arr_length = []
for id, polly, epi_chunk in polly_episode_arr:
    arr_length.append([id , len(polly), len(epi_chunk), polly.mean(), epi_chunk.mean(), polly.max(), epi_chunk.max(), polly.min(), epi_chunk.min()])
print(arr_length)
```

```python
import pandas as pd

audio_df = pd.DataFrame(arr_length, columns=['episode_id',
                                             'polly_length', 
                                             'episode_length',
                                             'polly_avg',
                                            'episode_avg',
                                            'polly_max_length',
                                            'episode_max_length',
                                            'polly_min_length',
                                            'episode_min_length'])
audio_df.head()
```

```python
#here I am comparing the waveforms of the polly generated audio and the episode audio chunks. The idea being
#that the same words should have similar waveforms
arr = np.array(polly_episode_arr)

for episode_id in arr[:5]:
    final_arr = np.delete(episode_id, episode_id[0][0], 0)
    poll_arr = final_arr[0]
    epi_arr = final_arr[1]
    plt.figure(figsize=(14, 5))
    fig, ax = plt.subplots(nrows=2, sharey=True)
    librosa.display.waveplot(y=poll_arr, sr=sr_pol, ax=ax[0])
    librosa.display.waveplot(y=epi_arr, sr=sr_episode, ax=ax[1])
    plt.show()
```

There seems to be some similarity between the polly generated speech and the podcast episode splices. Comparing the two, we can see the podcast episode has the general shape of the polly generated speech but the podcast has a lot more noise. This makes sense because there is a lot more noise in the podcast episode.

```python
#display Spectrogram of polly generate speech and podcast episode chunks
for spec in arr[:5]:
    final_spec = np.delete(spec, spec[0][0], 0)
    pol_spec = final_spec[0]
    epi_spec = final_spec[1]
    Y_pol = librosa.stft(pol_spec)
    Y_epi = librosa.stft(epi_spec)
    pol_db = librosa.amplitude_to_db(abs(Y_pol))
    epi_db = librosa.amplitude_to_db(abs(Y_epi))
    plt.figure(figsize=(14, 5))
    fig, ax = plt.subplots(nrows=2, sharey=True)
    librosa.display.specshow(pol_db, sr=sr_pol, x_axis='time', y_axis='hz', ax=ax[0])
    librosa.display.specshow(epi_db, sr=sr_episode, x_axis='time', y_axis='hz', ax=ax[1])
    plt.show()
```

Comparing the spectrograms, the shape between the podcast splices and the polly generated speech are the same. However, we can see that the polly generated speech has lower frequencies than the podcast episode splices. This makes sense because the polly generated speech is only one voice and is pretty monotone compared to the podcast episode splices.

```python
mfccs = librosa.feature.mfcc(y_pol, sr=sr_pol)
print(mfccs.shape)#Displaying  the MFCCs:
librosa.display.specshow(mfccs, sr=sr_pol, x_axis='time')
plt.ylabel('MFCCs')
plt.colorbar()
plt.title("MFCCs for Polly generated words")
```

```python
mfccs_epi = librosa.feature.mfcc(y_episode, sr=sr_episode)
print(mfccs_epi.shape)
librosa.display.specshow(mfccs_epi, sr=sr_episode, x_axis='time')
plt.ylabel('MFCCs')
plt.colorbar()
plt.title(" MFCCS for podcast episode")
```

Mel Frequency Cepstral Coefficient or MFCC plots, the words have about the same energy as each other. MFCCs are a special feature that captures the textural aspects of sound. They are unique because they appromixate to the human auditory system. In other words, the MFCC are able to capture of the important features of words and speech that humans are able to. This might become very important when I begin model building.
